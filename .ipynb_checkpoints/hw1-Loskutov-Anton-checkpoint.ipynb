# Туториал: Имплементация собственного метода

Для решения задач вида "имплементируйте метод X" крайне желательно (но необязательно) реализовывать методы с помощью интерфейса sklearn. Ниже вы увидите пример очень простого **классификатора**. Используйте эти наработи для выполнения задания.

Загрузите [датасет](https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv) с информацией о характеристиках вина и его качестве. (В метод `pd.read_csv()` можно подать веб-ссылку на данные и пандас их скачает и преобразует в `DataFrame`).

```python
df_wine = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv', sep=';')
```

```python
df_wine.head()
```

Наша задача состоит в предсказании целевого признака `quality` по остальным признакам. `quality` - это вещественная переменная (оценка качества от 0 до 10). Упростим себе задачу - создадим новый столбец `quality_cat` по следующим правилам:

* `quality_cat = 1` если `quality > 5` 
* `quality_cat = 0` - иначе 

Таким образом, `quality_cat` характерует "плохое" и "хорошее" вино.

```python
def prepocess_wine(df):
    df.loc[:, 'quality_cat'] = (df.loc[:, 'quality'] > 5).astype(int)
    df = df.drop('quality', axis=1)
    
    return df
```

```python
df_wine = df_wine.pipe(prepocess_wine)
df_wine.head()
```

Нарисуем распределение хорошего и плохого вина по признаку `alcohol`.

```python
_, ax = plt.subplots(1,1)
df_wine.query('quality_cat == 1').hist('alcohol', alpha=0.3, bins=20, label='good', ax=ax)
df_wine.query('quality_cat == 0').hist('alcohol', alpha=0.3, bins=20, label='bad', ax=ax)
ax.set_xlabel('alcohol')
ax.set_ylabel('count')
plt.legend()
```

Визуально кажется, что можно найти некоторое пороговое значение признака `alcohol` $\theta$, чтобы хорошо дискриминировать между плохими и хорошим вином:
* Если `alcohol >` $\theta$ - то предсказываем, что вино хорошее
* Если `alcohol <=` $\theta$ - то предсказываем, что вино плохое

Обучаеть классификатор будем очень просто: перебирать значения признака alchol и считать **ошибку** классификации. То значение, где ошибка минимальна и будет оптимальным.

Разделим целевой признак от остальных признаков

```python
X, y = df_wine.iloc[:, :-1].values, df_wine.iloc[:, -1].values
```

Далее вы увидете пример реализации этого простого классификатора.

```python
from sklearn.base import BaseEstimator, ClassifierMixin

class SimpleClassifier(BaseEstimator, ClassifierMixin): 
    
    def __init__(self):
        """
        Инициализация метода
        """
        
        self.threshold = 0

    def fit(self, X, y=None):
        """
        Выполняет процедуру обучения классификатора. В данном случае - перебор занчения признака alcohol.
        Договоримся, что признак alcohol будет в последнем столбце матрицы X
        """
        
        alcohol = X[:,-1]
        n = float(y.shape[0])
        
        # Задаем пространство поиска
        alco_range = np.arange(alcohol.min(), alcohol.max(), 0.1)
        
        min_error_rate = 1
                
        #  Начинаем поиск
        for theta in alco_range:
            prediction = alcohol > theta
            
            error_rate = sum(prediction != y)/n
            
            if error_rate < min_error_rate:
                min_error_rate = error_rate
                self.threshold = theta  
        
        return self

    def predict(self, X, y=None):
        """
        Процедура предсказания класса по обученному порогу
        """
        
        return (X[:,-1] > self.threshold).astype(int)
```

Обучаем модель

```python
model = SimpleClassifier()
model.fit(X, y)

print 'Оптимальный порог: %f' % model.threshold
```

Получаем предсказание модели (на обучающей выборке)

```python
y_hat = model.predict(X)
```

Оценим долю ошибок на обучающей выборке

```python
error_rate = sum(y_hat != y)/float(y.shape[0])
print 'Доля ошибок на обучающей выборке: %f' % error_rate
```

```python
import copy
from sklearn.metrics.pairwise import euclidean_distances
```

```python
from sklearn.base import ClusterMixin
from scipy.spatial import distance
import math

class Kmeans(BaseEstimator, ClusterMixin): 
    
    def __init__(self, k=2, metric='euclidean', max_iter=100, random_state=None, init='random'):
        """
        Инициализация метода
        :k - количество кластеров
        :metric - функция расстояния между объектами
        :max_iter - максиальное количество итераций
        :random_state - seed для инициализации генератора случайных чисел
        """
        
        self.k = k
        self.random_state = random_state
        self.metric = metric
        self.max_iter = max_iter
        self.init = init

    def fit(self, X, y=None):
        """
        Процедура обучения k-means
        """
        
        self.labels = np.empty(X.shape[0], dtype = np.uint8)
        self.centroids = np.empty((self.k, X.shape[1]))
        self.centroids[0] = np.nan
        
        ## Your Code Here
        while not (np.isfinite(self.centroids).all()):
            if (self.init == 'k-means'):
                self.centroids[0] = X[np.random.randint(0, self.labels.size-1)]
            
                for n in xrange(1, self.k):
                    dist = euclidean_distances(X, self.centroids[:n])
                    dist_sum = dist.min(1).sum()
                    random_sum = np.random.rand() * dist_sum
                    dist_sum = 0
                    for i in xrange(X.shape[0]):
                        dist_sum += dist[i][n-1]
                        if dist_sum > random_sum:
                            self.centroids[n] = X[i]
                            break
            else:
                self.centroids = X[np.random.randint(0, self.labels.size-1, self.k)]
        
        for i in range(self.max_iter):
            dists = np.array(euclidean_distances(X, self.centroids))
            self.labels = dists.argmin(1)
            for k in range(self.k):
                ind = np.where(self.labels == k)[0]
                if (ind.size != 0):
                    new_centroid = np.vstack([X[ind].mean(0)])
                    if np.isfinite(new_centroid).all():
                        self.centroids[k] = new_centroid
            

    def predict(self, X, y=None):
        """
        Процедура предсказания кластера
        
        Возвращает метку ближайшего кластера для каждого объекта
        """
        return self.labels

class MiniBatchKMeans(Kmeans):
    
    def __init__(self, k=2, metric='euclidean', max_iter=100, random_state=None, init='random', batch_size=50):
        """
        Инициализация метода
        :k - количество кластеров
        :metric - функция расстояния между объектами
        :max_iter - максиальное количество итераций
        :random_state - seed для инициализации генератора случайных чисел
        :batch_size - Размер батча для MiniBatchKMeans
        """
        
        self.k = k
        self.random_state = random_state
        self.metric = metric
        self.max_iter = max_iter
        self.batch_size = batch_size
        self.init = init
        
    def fit(self, X, y=None):
        """
        Процедура обучения MiniBatch k-means
        """
        
        if X.shape[0] <= self.batch_size :
            return Kmeans(k=self.k, max_iter=self.max_iter, init=self.init).fit(X, y)
        
        self.labels = np.empty(X.shape[0], dtype = np.uint8)
        self.centroids = np.empty((self.k, X.shape[1]))
        
        if (self.init == 'k-means'):
            self.centroids[0] = X[np.random.randint(0, self.labels.size-1)]
            
            for n in xrange(1, self.k):
                dist = euclidean_distances(X, self.centroids[:n])
                dist_sum = dist.min(1).sum()
                random_sum = np.random.rand() * dist_sum
                dist_sum = 0
                for i in xrange(X.shape[0]):
                    dist_sum += dist[i][n-1]
                    if dist_sum > random_sum:
                        self.centroids[n] = X[i]
                        break
        else:
            self.centroids = X[np.random.randint(0, self.labels.size-1, self.k)]

        for m in range(self.max_iter):
            batch = X[np.random.randint(0, X.shape[0], self.batch_size)]
            self.labels = euclidean_distances(batch, self.centroids).argmin(1)
            
            self.v = np.zeros(self.k)
            for i in xrange(self.k):
                ind = np.where(self.labels == i)[0]
                self.v[i] = 1./(ind.sum()+1)
                
            for i in xrange(self.batch_size):
                tmp = self.labels[i]
                n = self.v[tmp]
                self.centroids[tmp] = (1-n)*self.centroids[tmp] + n*batch[i]
                
    def predict(self, X, y=None):
        """
        Процедура предсказания кластера
        
        Возвращает метку ближайшего кластера для каждого объекта
        """
        
        return euclidean_distances(X, self.centroids).argmin(1)
```

# Вывод
По построенным графикам выше видно, что алгоритмы k-means и k-means++ не сильно отличаются в скорости, скорее всего из-за того, что алгоритмы выполнялись заданное(max_iter) число раз и не выходили, использую некий эпсилон. Но алгоритм MiniBatch k-means ощутимо быстрее простой реализации k-means.

Из построенных графиков кластеров четко выделяются визуально кластера и признаки, которые в них выделил k-means, например, в 1 это большая любовь к волейболу, а в 5 к баскетболу. Также есть "усредняющие кластеры", люди в которых не проявляют большую любовь к какому-то одному или паре увлечений, такие как 2 и 3 кластер например.
В среднем кластеры можно описать так:
1 - Люди, проявляющие интерес к волейболу
2 - Люди, проявляющие интерес к спорту в общем(немного) и в остальном в среднем немного выше, чем у остальных, но в среднем как остальные
3 - Люди, проявляющие интерес к покупкам и красоте, но в среднем их интересы такие же как и у всех остальных
4 - Люди, проявляющие интерес к вере
5 - Люди, проявляющие интерес к баскетболу
6 - Люди, интересы которых в среднем ниже, чем у остальной выборки
7 - Люди, проявляющие интерес к hollister и abercrombie
8 - Люди, проявляющие интерес к походам и состоящие в группах
9 - Люди, проявляющие интерес к поцелуям и прическе, также к полу, но в среднем более активные, чем остальная выборка
