Алгоритм кластеризации должен удовлетворять следующему интерфейсу. Конструктор принимает набор параметров, необходимых для работы алгоритма кластеризации. Метод `fit` подсчитывает параметры модели и возвращает `self`. Метод `predict` возвращает вектор с индексами кластеров для поданных в него объектов `x`

```python
## Задание 1
from sklearn.base import BaseEstimator, ClusterMixin

class Clustering(BaseEstimator, ClusterMixin):
    """
    Implement clustering algorithm according 
    """
    
    def __init__(self, epsilon=0.7, min_pts=5):
        """
        Please add necessary algoritm parameters to class constructor.
        """
        self.epsilon = epsilon
        self.min_pts = min_pts
        self.C = 0
        return
     
    def find_neighbours(self, p):
        """
        Ищем ближайщих к точке p
        """
        return np.array([i for i in xrange(self.x_len) if self.dists[p][i] < self.epsilon])
    
    def expand_cluster(self, neighbours, p):
        """
        Пытаемся дополнить кластер self.C точкой p и ее ближайщими и тд
        """
        self.X[p][-1] = self.C
        while len(neighbours) > 0: 
            p = neighbours[0]
            neighbours = np.delete(neighbours, 0)
            if self.X[p][-2] == 1:
                continue
            else:
                self.X[p][-2] = 1
            new_neighbours = self.find_neighbours(p)
            if len(new_neighbours) >= self.min_pts:
                neighbours = np.append(neighbours, new_neighbours)
            if self.X[p][-1] <= 0:
                self.X[p][-1] = self.C
    
    def fit(self, X, y=None):
        """
        Use data matrix x to compute model parameters
        """
        self.X = X
        self.x_len = len(self.X)
        self.dists = ss.distance.cdist(self.X, self.X)
        
        # Добавляем столбец для определения посещенных точек(0 - не посетили, 1 - посетили)
        self.X = np.hstack((self.X, np.zeros((self.X.shape[0], 1))))

        # Добавляем столбец, в который будем заносить номер кластера
        self.X = np.hstack((self.X, np.zeros((self.X.shape[0], 1))))
        
        for i in xrange(self.x_len):
            if self.X[i][-2] == 1:
                continue
            self.X[i][-2] = 1
            neighbours = self.find_neighbours(i)
            if len(neighbours) < self.min_pts:
                X[i][-1] = -1
            else:
                self.C += 1
                self.expand_cluster(neighbours, i)
        
        return self
    
    def predict(self, x):
        """
        Using computed model parameters predict cluster
        for all objects from x
        """
        return self.X[:,-1]
    
    def fit_predict(self, x, y=None):
        self.fit(x, y)
        return self.predict(x)
    
```

Выбрать и реализовать критерий качества, на основании которого будут подбираться параметры модели. Варианты критериев:
- Инерция (сумма квадратов расстояний от каждой из точек до ближайшего к этой точке центроида)
- Средний диаметр (максимальное расстояние между точками в кластере) или радиус (расстояние от центроида до самой дальней от него точки в кластере)
- Sihouette

**Критерий качества необходимо выбрать таким образом, чтобы он подходил для реализованного алгоритма**

```python
## Задание 2
def quality(X, y):
    """
    Реализует критерий качества Silhouette
    """
    dist = ss.distance.cdist(X, X)
    unique, counts = numpy.unique(y, return_counts=True)
    #n_labels= len(unique)
    
    # A - расстояние от точки до своего кластера
    A = np.zeros(X.shape[0])
    # B - расстояние от точки до ближайщего чужого кластера
    B = np.full(X.shape[0], np.inf)
    
    for i, k in enumerate(unique):
        dist_for_point = dist[y == k]
        A[y == k] = np.sum(dist_for_point[:, y == k], axis=1)
        if counts[i] != 0 and np.isnan(counts[i]):
            A[y == k] /= counts[i]
        for j, l in enumerate(unique):
            if l != k:
                dist_to_other = np.mean(dist_for_point[:, y == l], axis=1)
                B[y == k] = np.minimum(B[y == k], dist_to_other)
    return (B - A) / np.maximum(A, B)
```

Применим критерий для выбора параметров алгоритма. Предлагается изменить следующий код таким образом, чтобы кластеризация вызывалась с верным набором параметров. На графике можно будет увидеть зависимость критерия от параметров алгоритма. Необходимо выбрать оптимальные значения этих параметров.

```python
ks = range(1, 21)
criteria = np.zeros(len(ks))
x = x_iris

for i, k in enumerate(ks):
    cls = Clustering(epsilon=0.7, min_pts=k)
    y = cls.fit_predict(x)
    criteria[i] = quality(x, y).mean()
    
pl.figure(figsize=(8, 6))
pl.plot(ks, criteria, marker='o')
pl.title("$J(k)$")
pl.ylabel("Criteria $J$")
pl.xlabel("Number of minimum near points")
pl.grid()
pl.show()
```
